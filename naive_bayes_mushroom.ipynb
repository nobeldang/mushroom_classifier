```python
import numpy as np
#Preprocessing:
from sklearn.preprocessing import LabelEncoder
#for splitting training and test data
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
df=pd.read_csv('./Datasets/mushrooms.csv')
df.head()
```

```python
print df.shape
```

```python
le=LabelEncoder()
ds=df.apply(le.fit_transform)
```

```python
ds.head()
```

```python
data = ds.values
print data.shape
```

```python
data_x=data[:,1:]
data_y=data[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_x,data_y,test_size=0.2)
print x_test.shape
print y_test.shape
```

```python
def prior_prob(y_train,label):
    total=y_train.shape[0]
    labelled=np.sum(y_train==label)
    return labelled/float(total)
```

```python
def conditional_prob(x_train,y_train,feature_col,feature_value,label):
    denom=np.sum(y_train==label)
    x_filtered=x_train[y_train==label]
    numer=np.sum(x_filtered[:,feature_col]==feature_value)
    
    return numer/float(denom)
```

```python
def predict(x_train,y_train,xtest):
    classes=np.unique(y_train)
    n_features=x_train.shape[1]
    post_probs=[]
    for label in classes:
        likelihood=1
        for f in range(n_features):
            cond=conditional_prob(x_train,y_train,f,xtest[f],label)
            likelihood*=cond
        prior=prior_prob(y_train,label)  
        post=likelihood*prior
        post_probs.append(post)
    pred=np.argmax(post_probs)  
    return pred
```

```python
def score(x_train,y_train,x_test,y_test):
    pred=[]
    for i in range(x_test.shape[0]):
        pred_label=predict(x_train,y_train,x_test[i])
        pred.append(pred_label)
    pred=np.array(pred) 
    accuracy=float(np.sum(pred==y_test))/y_test.shape[0]
    return accuracy
```

```python
print score(x_train,y_train,x_test,y_test)
```
